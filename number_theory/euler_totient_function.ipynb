```python

def gcd(a,b):
    while b != 0:
        a, b = b, a % b
    return a

def coprime(a,b):
    return gcd(a,b) == 1

totient_memo = {}

def totient(x):
    if x in totient_memo:
        return totient_memo[x]
    ans = 0
    for i in range(1,x):
        if coprime(i,x):
            ans += 1
    totient_memo[x] = ans
    return ans
```

```python
i_max = 1e4
for i in range(1,int(i_max)+1):
    totient(i)
```

```python
def color_1_100(x):
    assert x >= 0
    assert x <= 100
    r = min(x,33)
    g = max(0,min(33,x-33))
    b = max(0,min(33,x-66))

    def normalize(v):
        return int(255/33 * (33-v))

    rgb_tuple = (normalize(r),normalize(g),normalize(b))
    return '#%02x%02x%02x' % rgb_tuple
```

```python
import matplotlib.pyplot as plt
import scienceplots
import numpy as np

x_max = 1e3

x = list(range(1,int(x_max)+1))
y = [totient(xi) for xi in x]
# colors = [color_1_100(xi/x_max * 100) for xi in x]

with plt.style.context(["science"]):

    fig, ax = plt.subplots(figsize = (8,5))

    plt.plot(x,y,linewidth=0.5)
    plt.ylabel("$\phi(x)$")
    plt.xlabel("$x$")
    plt.show()
```

```python
import matplotlib.pyplot as plt
import scienceplots
import numpy as np

x_max = 1e4

x = list(range(1,int(x_max)+1))
curr_value = 0
y = []
for xi in x:
    if totient(xi) == xi-1:
        curr_value += 1
        y.append(curr_value)
    else:
        y.append(curr_value)

y2 = []
curr_value = 0
for xi in x:
    if totient(xi) == xi/2-1:
        curr_value += 1
        y2.append(curr_value)
    else:
        y2.append(curr_value)


with plt.style.context(["science"]):

    fig, ax = plt.subplots(figsize = (8,5))

    plt.plot(x,y)
    plt.plot(x,y2)
    plt.show()
```
